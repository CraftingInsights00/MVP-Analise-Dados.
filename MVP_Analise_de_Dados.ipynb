# **Validando o Encoding dos Arquivos CSV**
Antes de consolidar as bases anuais, é importante garantir que a leitura dos arquivos preserve corretamente os caracteres especiais (acentuação). O trecho abaixo lê apenas as duas primeiras linhas de cada arquivo, utilizando o encoding "latin1", e exibe exemplos de colunas textuais onde normalmente ocorrem problemas, para conferência visual.

```python
for arq in arquivos:
    print(f'Arquivo testado: {arq}')  # Indica qual arquivo está sendo testado.
    try:
        # Lê apenas as duas primeiras linhas do arquivo, utilizando o encoding "latin1"
        df_sample = pd.read_csv(arq, sep=';', encoding='latin1', nrows=2)
        # Mostra as colunas que podem conter caracteres especiais
        print(df_sample[['municipio', 'causa_acidente', 'tipo_acidente']].head())
    except Exception as e:
        # Em caso de erro, imprime uma mensagem detalhando o problema
        print(f"Erro ao ler {arq}: {e}")
    print('-'*40)  # Separador para facilitar a visualização dos resultados
```

# **Validação das Colunas dos Arquivos CSV**
Para cada arquivo CSV, lemos apenas o cabeçalho (nomes das colunas) e salvamos para confronto posterior.

```python
colunas_por_arquivo = {}
for arq in arquivos:
    try:
        # Lê só o cabeçalho, sem carregar dados, usando encoding já validado
        df_temp = pd.read_csv(arq, sep=';', encoding='latin1', nrows=0)
        colunas_por_arquivo[arq] = list(df_temp.columns)
        print(f"Colunas presentes em {arq.split('/')[-1]}:\n{list(df_temp.columns)}\n")
    except Exception as e:
        print(f"Erro ao ler o cabeçalho de {arq}: {e}")
```

Após coletar os cabeçalhos, checamos se todas as colunas são idênticas entre os arquivos.



```python
# Checando se todas as colunas são idênticas entre os arquivos
conjuntos_colunas = [set(cols) for cols in colunas_por_arquivo.values()]
schemas_iguais = all(conjuntos_colunas[0] == c for c in conjuntos_colunas[1:])
if schemas_iguais:
    print("\n✅ Todos os arquivos possuem o mesmo conjunto de colunas.")
else:
    print("\n⚠️ Atenção: Foram encontradas diferenças nos nomes de colunas. Veja acima para análise detalhada.")
```

# **Armazenamento e Consolidação dos DataFrames Anuais**

Nesta etapa, armazenamos cada DataFrame anual já com a coluna 'ano' para facilidade na análise posterior.

```python
dfs = []
# Loop por cada arquivo já validado
for idx, arq in enumerate(arquivos):
    try:
        df = pd.read_csv(arq, sep=';', encoding='latin1', low_memory=False)
        df['ano'] = anos[idx]  # Adiciona coluna do ano, baseada na ordem dos arquivos/anos
        dfs.append(df)
        print(f"Arquivo {arq.split('/')[-1]} lido com sucesso: {df.shape[0]} linhas, {df.shape[1]} colunas.")
    except Exception as e:
        print(f"Erro ao ler {arq}: {e}")
```

# **Razão da Consolidação dos DataFrames**
Unimos os DataFrames anuais em um único DataFrame consolidado para facilitar a análise. Isso centraliza todos os dados de acidentes de trânsito em uma única estrutura, permitindo operações de análise e visualização mais eficientes. O parâmetro ignore_index=True garante que o índice seja reorganizado, evitando conflitos e mantendo a integridade dos dados.

```python
# Concatenação dos DataFrames em um único DataFrame consolidado
df_consolidado = pd.concat(dfs, ignore_index=True)
print("\nConsolidação concluída.")
print(f"DataFrame único de acidentes de trânsito - Linhas: {df_consolidado.shape[0]}; Colunas: {df_consolidado.shape[1]}\n")
# Visualizamos as 5 primeiras linhas para garantir a estrutura
df_consolidado.head()
```

Nesta fase, verificamos a presença de linhas duplicadas no DataFrame consolidado e, se encontrado, removemos as duplicatas.

```python
# Checagem e eventual remoção de linhas duplicadas
duplicados = df_consolidado.duplicated()
n_duplicados = duplicados.sum()
print(f"Total de linhas duplicadas (todas as colunas idênticas): {n_duplicados}")

# Função para caso haja duplicação, utilizarmos somente a primeira ocorrencia
if n_duplicados > 0:
    df_consolidado = df_consolidado[~duplicados]
    print(f"Linhas após remoção dos duplicados: {df_consolidado.shape[0]}")
```

# **Conversão e Padronização de Dados**
Após a análise inicial, identificamos a necessidade de conversão e padronização de algumas colunas essenciais:

* **Latitude e Longitude:** As colunas de latitude e longitude estão armazenadas como objetos (strings), o que pode dificultar análises geoespaciais. Precisamos convertê-las para o tipo numérico, garantindo que sejam tratadas corretamente em análises posteriores.
* **Coluna 'km':** Assim como as coordenadas, a coluna 'km' também é fundamental para entender as distâncias associadas aos acidentes. A conversão para o tipo numérico permite cálculos precisos e evita erros.
* **Formato datetime:** A coluna 'data_inversa' deve ser convertida para o formato datetime. Isso nos permitirá realizar análises temporais de maneira mais eficiente, como identificar tendências ao longo do tempo e agrupar dados por períodos.

```python
# Conversão e padronização de latitude e longitude
df_consolidado['latitude'] = df_consolidado['latitude'].astype(str).str.replace(',', '.', regex=False)
df_consolidado['latitude'] = pd.to_numeric(df_consolidado['latitude'], errors='coerce')
df_consolidado['longitude'] = df_consolidado['longitude'].astype(str).str.replace(',', '.', regex=False)
df_consolidado['longitude'] = pd.to_numeric(df_consolidado['longitude'], errors='coerce')

# Conversão e padronização da coluna km
df_consolidado['km'] = df_consolidado['km'].astype(str).str.replace(',', '.', regex=False)
df_consolidado['km'] = pd.to_numeric(df_consolidado['km'], errors='coerce')

# Conversão de data para formato datetime
df_consolidado['data_inversa'] = pd.to_datetime(df_consolidado['data_inversa'], errors='coerce')
```

# **Motivação para a Filtragem Geográfica**
Com base na identificação de valores extremos e inconsistentes nas coordenadas, decidimos implementar um filtro rigoroso. O objetivo é remover quaisquer registros com dados geográficos inválidos, garantindo que o DataFrame resultante represente fielmente os acidentes ocorridos dentro dos limites territoriais brasileiros.

**Ações Realizadas:**
1. **Definição de Intervalos Válidos:** Estabelecemos limites aceitáveis para latitude e longitude, baseando-nos em parâmetros geográficos conhecidos.
2.  **Contabilização de Registros:** Antes de aplicar o filtro, contabilizamos o número total de registros para garantir a rastreabilidade das alterações.
3.   **Aplicação do Filtro:** Implementamos a filtragem, excluindo registros com coordenadas fora dos intervalos definidos. Essa ação é crucial para assegurar que os dados restantes sejam consistentes, aumentando a confiabilidade para as etapas futuras de análise.
4. **Documentação da Remoção dos Registros:** Por fim, documentamos a quantidade de registros removidos, permitindo uma análise crítica sobre a quantidade de dados que pode ter sido afetada por inconsistências.

# **Considerações sobre Práticas de Negócio**
Em um ambiente de trabalho ideal, a tomada de decisões sobre dados deve ser baseada em colaboração com profissionais de negócios ou stakeholders. Ao identificar dados inconsistentes, seria fundamental consultar as partes interessadas para discutir as melhores abordagens quanto a esses dados. Embora a exclusão de registros errôneos geralmente seja uma prática recomendada para garantir a integridade da análise, este processo deve ser validado com a equipe de negócios. Eles podem oferecer insights valiosos sobre o impacto e a relevância desses dados, além de sugerir alternativas viáveis, como a correção dos dados, se possível.

Neste caso, optamos pela exclusão dos registros inconsistentes por ser a abordagem mais direta e segura, assegurando que a análise subsequente seria realizada sobre uma base de dados precisa e confiável. No entanto, essa decisão reflete um entendimento da situação atual e está sujeita à validação em contextos reais de trabalho, onde a colaboração e a comunicação são essenciais.

```python
# Definição dos intervalos válidos para as coordenadas brasileiras
faixa_latitude = (-34, 6)
faixa_longitude = (-74, -34)

# Contabilização do total de registros antes da filtragem geográfica
total_antes_geo = len(df_consolidado)

# Aplicação do filtro para remoção de registros com coordenadas inconsistentes com o território brasileiro
df_consolidado = df_consolidado[
    (df_consolidado['latitude'].between(*faixa_latitude)) &
    (df_consolidado['longitude'].between(*faixa_longitude))
]

# Contabilização do total de registros pós-filtragem geográfica
total_depois_geo = len(df_consolidado)

# Documentação da quantidade de registros removidos por esse critério
print(f"Registros removidos por coordenadas fora dos limites geográficos do Brasil: {total_antes_geo - total_depois_geo}")
print(f"Total de registros após filtragem geográfica: {total_depois_geo}")

# Visualização das primeiras linhas, já com dados consistentes de latitude, longitude e km
display(df_consolidado[['data_inversa', 'horario', 'latitude', 'longitude', 'km']].head())
```

# **Identificação de Ausências Disfarçadas**
Com o intuito de garantir a qualidade dos dados, implementamos uma função que busca possíveis “ausências disfarçadas” em colunas específicas. Esses são valores que, embora não estejam tecnicamente ausentes, podem representar informações faltantes, como strings como 'N/A', 'Null', ou 'Sem Info'.

**Função para Busca de Ausências Disfarçadas**

```python
# Função para busca de ausências disfarçadas em uma coluna
def busca_ausencias(col):
    col_lower = df_consolidado[col].astype(str).str.lower()  # Converte os valores da coluna para minúsculas
    return col_lower.value_counts().loc[
        col_lower.value_counts().index.str.contains('n/a|na|null|indefi|ignora|sem info|vazio|-', case=False)  # Filtra possíveis entradas de ausência
    ].sort_values(ascending=False)  # Ordena os resultados
```

A execução da função de busca por ausências disfarçadas ocorreu conforme o esperado e não foram identificados casos desse tipo nas colunas analisadas. O código foi executado com sucesso, sem erros, validando a consistência dos dados nesse aspecto.

```python
# Checagem nas colunas principais com ausentes
for coluna in ['uop', 'br', 'km']:
    print(f"\nPrincipais 'ausências disfarçadas' em {coluna}:")
    print(busca_ausencias(coluna))  # Imprime os resultados da análise
```

# **Processo de Padronização**
O processo de padronização inclui as seguintes etapas:

1. **Remoção de Espaços:** Removemos espaços em branco antes e depois dos valores.

2. **Substituição de Espaços Duplos:** Substituímos quaisquer espaços duplos por um único espaço.

# **Verificação de Possíveis Duplicidades nos Nomes dos Municípios**
Após a contagem dos acidentes, é importante assegurar que os dados sobre municípios sejam consistentes. Nomes de municípios que podem ser semanticamente semelhantes, como "SAO PAULO", "SÃO PAULO" e "SAO PAULO", podem resultar em contagens duplicadas e distorcer a análise. Para tratar isso, usamos a biblioteca difflib para identificar nomes com alta similaridade.

```python
municipios_unicos = municipios_freq['municipio'].tolist()
possiveis_dupes = []

# Checa pares com alta similaridade (threshold pode ser ajustado)
for i, m1 in enumerate(municipios_unicos):
    similares = difflib.get_close_matches(m1, municipios_unicos, n=5, cutoff=0.92)
    for m2 in similares:
        if m1 != m2 and (m2, m1) not in possiveis_dupes:  # Evita duplicidade reversa
            possiveis_dupes.append((m1, m2))
```

Em seguida, exibimos uma amostra das possíveis duplicidades encontradas:

```python
# Exibe as possíveis duplicidades (amostra se a lista ficar muito longa)
print("Amostra de possíveis duplicidades semânticas (alta similaridade de nome):")
print(possiveis_dupes[:20])
```

A verificação de possíveis duplicidades semânticas entre nomes de municípios não identificou inconsistências relevantes na base de dados. # **Análise de Horários**
Após verificadas as datas, verificamos se os horários estão no formato adequado e se seguem os padrões esperados.

```python
# Diagnóstico de horários em formato inesperado
def horario_invalido(horario):
    # Esperado: HH:MM:SS ou HH:MM, com horas/minutos válidos
    match = re.match(r'^\d{2}:\d{2}(:\d{2})?$', horario)
    if not match:
        return True
    try:
        h, m, *s = [int(x) for x in horario.split(':')]
        if not (0 <= h < 24 and 0 <= m < 60):
            return True
        if s and not (0 <= s[0] < 60):
            return True
        return False
    except:
        return True

horas_unicas = df_consolidado['horario'].dropna().unique()  # Coleta horários únicos não nulos
horas_invalidas = [h for h in horas_unicas if horario_invalido(h)]  # Filtra horários inválidos
print(f"Quantidade de formatos de hora inválidos/fora do padrão: {len(horas_invalidas)}")
print("Exemplos de horários inválidos:")
print(horas_invalidas[:10])  # Exibe exemplos de horários inválidos
```

A verificação confirmou que todos os horários estão no formato adequado e seguem o padrão esperado, não havendo registros com formatação incorreta ou valores inválidos nesse campo. Os dados de horário estão prontos para análises temporais mais detalhadas.

Durante a análise gráfica das variáveis relacionadas a vítimas e pessoas envolvidas nos acidentes, observei a presença de valores extremos (outliers) especialmente nas variáveis que indicam o número de mortos, feridos leves, feridos graves e total de pessoas envolvidas por tipo de acidente. Esses outliers se destacam por apresentar quantidades muito superiores à média, o que pode indicar situações atípicas, possíveis inconsistências no preenchimento dos dados ou, de fato, eventos de grandes proporções.

A presença desses valores extremos pode distorcer a interpretação dos padrões gerais do conjunto de dados, dificultando a identificação fiel dos tipos de acidentes mais críticos sob uma ótica de frequência e severidade. Por exemplo, boxplots exibiram, em alguns tipos de acidente, pontos muito distantes da maioria, sugerindo eventos com número incomum de mortos ou vítimas.

Diante desse cenário, torna-se fundamental avaliar o impacto desses outliers sobre as estatísticas agregadas e sobre eventuais decisões analíticas. Este processo de investigação não implica necessariamente a exclusão dos extremos, mas sim uma análise crítica desses casos para compreender sua natureza, frequência e influência sobre o todo.

Por isso, será realizada uma análise complementar para mensurar o peso desses outliers nas principais variáveis, identificar como eles afetam médias e totais, e avaliar se sua presença compromete a robustez das conclusões. Assim, demonstramos um zelo metodológico importante, evidenciando que a variável foi revisada e que decisões sobre tratamento ou não desses casos seriam tomadas em conjunto com o responsável do negócio, de acordo com o contexto específico da análise.

```python
def resumo_outliers_analitico(df, col, grupo=None, pct_outlier=1.5):
    if grupo is not None:
        resultados = []
        grupos = df.groupby(grupo)
        for nome, dados in grupos:
            serie = dados[col]
            # Considera apenas registros com valor > 0 no cálculo dos limites
            serie_validos = serie[serie > 0]
            if len(serie_validos) < 10:
                # Resultados não são robustos para grupos muito pequenos
                continue
            q1 = serie_validos.quantile(0.25)
            q3 = serie_validos.quantile(0.75)
            iqr = q3 - q1
            limite_sup = q3 + pct_outlier * iqr

            outliers = serie[serie > limite_sup]
            n_out = outliers.shape[0]
            total = serie_validos.shape[0]
            soma_out = outliers.sum()
            soma_total = serie_validos.sum()
            prop_soma = (soma_out / soma_total * 100) if soma_total > 0 else 0
            resumo = {
                'Grupo': nome,
                'N_casos_validos': total,
                'Mediana': serie_validos.median(),
                'Max': serie_validos.max(),
                'Percentil_95': serie_validos.quantile(0.95),
                'Moda': mode(serie_validos, keepdims=True).mode[0],
                'Limite_outlier': limite_sup,
                'Qtd_outliers': n_out,
                'Soma_outliers': soma_out,
                'Prop_soma_outliers_%': prop_soma,
            }
            resultados.append(resumo)
        return pd.DataFrame(resultados).sort_values('Prop_soma_outliers_%', ascending=False)
    else:
        # Aplicação no geral
        serie = df[col]
        serie_validos = serie[serie > 0]
        q1 = serie_validos.quantile(0.25)
        q3 = serie_validos.quantile(0.75)
        iqr = q3 - q1
        limite_sup = q3 + pct_outlier * iqr
        outliers = serie[serie > limite_sup]
        n_out = outliers.shape[0]
        total = serie_validos.shape[0]
        soma_out = outliers.sum()
        soma_total = serie_validos.sum()
        prop_soma = (soma_out / soma_total * 100) if soma_total > 0 else 0
        resumo = {
            'N_casos_validos': total,
            'Mediana': serie_validos.median(),
            'Max': serie_validos.max(),
            'Percentil_95': serie_validos.quantile(0.95),
            'Moda': mode(serie_validos, keepdims=True).mode[0],
            'Limite_outlier': limite_sup,
            'Qtd_outliers': n_out,
            'Soma_outliers': soma_out,
            'Prop_soma_outliers_%': prop_soma,
        }
        return pd.DataFrame([resumo])

# Aplicação por tipo de acidente:
variaveis = ['mortos', 'feridos_leves', 'feridos_graves', 'pessoas']
for v in variaveis:
    print(f"\nResumo analítico para {v.upper()} por tipo de acidente:\n")
    print(resumo_outliers_analitico(df_consolidado, v, grupo='tipo_acidente').head(10))
```

Após a análise dos outliers, elaboramos uma modelagem estatística para investigar quais fatores estão mais associados à gravidade dos acidentes no dataset. O próximo bloco de código realiza uma regressão logística, considerando como variável de interesse se o acidente resultou em morte ou feridos graves. Com isso, buscamos quantificar o efeito de variáveis contextuais — como dia da semana, condição meteorológica, fase do dia e tipo de pista — sobre a probabilidade de ocorrência de acidentes graves ou fatais. O resultado dessa análise permitirá identificar cenários de maior risco e orientar recomendações focadas em prevenção.

```python
# Criação da coluna binária para acidentes graves/fatais (mortos OU feridos graves)
df_consolidado['grave_fatal_1_0'] = df_consolidado.apply(
    lambda row: 1 if (row['mortos'] > 0 or row['feridos_graves'] > 0) else 0, axis=1
)

variaveis = ['dia_semana', 'condicao_metereologica', 'fase_dia', 'tipo_pista']

df_model = df_consolidado[variaveis + ['grave_fatal_1_0']].dropna()
X = pd.get_dummies(df_model[variaveis], drop_first=True)
y = df_model['grave_fatal_1_0']

# Conversão para float para evitar erro statsmodels
X = sm.add_constant(X)
X = X.astype(float)
y = y.astype(float)

modelo_log = sm.Logit(y, X).fit()

print(modelo_log.summary())

# Para visualização dos odds ratios
odds_ratios = np.exp(modelo_log.params)
conf = modelo_log.conf_int()
conf['OR'] = odds_ratios
conf.columns = ['IC 2.5%', 'IC 97.5%', 'Odds Ratio']

conf_no_const = conf.iloc[1:]  # remove a constante

plt.figure(figsize=(8,8))
plt.errorbar(
    conf_no_const['Odds Ratio'], conf_no_const.index,
    xerr=[
        conf_no_const['Odds Ratio']-np.exp(conf_no_const['IC 2.5%']),
        np.exp(conf_no_const['IC 97.5%'])-conf_no_const['Odds Ratio']],
    fmt='o', color='navy', alpha=0.8)
plt.axvline(1, color='red', linestyle='--')
plt.title('Impacto das Variáveis na Chance de Acidente Grave/Fatal (Odds Ratio)')
plt.xlabel('Odds Ratio (↑ risco >1, ↓ risco <1)')
plt.ylabel('Variável')
plt.tight_layout()
plt.show()

print("\nOdds Ratio e Intervalo de Confiança das variáveis do modelo:")
print(conf_no_const.sort_values('Odds Ratio', ascending=False))
```

Os resultados do modelo logístico e do gráfico de odds ratios apontam quais condições associam-se a maior ou menor risco de acidentes graves ou fatais dentro deste conjunto de dados. Vale ressaltar, entretanto, a necessidade de interpretar esses achados à luz do contexto observado, sem perder de vista como a exposição (por exemplo, tempo ou distância trafegada em cada tipo de pista) influencia os resultados.

No caso de “tipo de pista”, observa-se que pistas simples apresentam odds ratio elevado. Isso indica que, dentro do nosso dataset, acidentes graves/fatais acontecem com mais frequência relativa nesse tipo de pista. Contudo, é importante destacar que a imensa maioria dos quilômetros rodados e, possivelmente, dos registros em nossa base ocorre em pistas simples — no Brasil, são amplamente majoritárias em relação às pistas duplas ou múltiplas. Portanto, o resultado não sugere, necessariamente, que pistas simples sejam inerentemente mais perigosas, mas pode refletir simplesmente o fato de que os motoristas passam muito mais tempo circulando por elas, elevando a exposição ao risco nesse ambiente.

O mesmo cuidado vale para as condições meteorológicas analisadas. Climas aparentemente “seguros”, como sol e céu claro, aparecem com maior associação à gravidade, possivelmente porque são também os cenários mais comuns de circulação, ou ainda por influência de comportamentos menos defensivos dos motoristas nessas situações. Não se deve interpretar isso como recomendação de dirigir sob chuva intensa, mas sim entender que, no conjunto de acidentes observados, grande parte deles ocorre naturalmente em condições “favoráveis”, pois são também as mais prevalentes.

De modo geral, o modelo evidencia que variáveis ambientais e do contexto rodoviário impactam os desfechos, mas a exposição e a frequência de uso de cada categoria precisam sempre ser considerados na análise e na interpretação dos resultados. Essa abordagem garante que intervenções e políticas públicas partam não só dos dados, mas de uma compreensão realista do cenário rodoviário.

Em conclusão, interpretações estatísticas só têm sentido pleno quando contextualizadas pelo uso real das rodovias e pela dinâmica do fluxo viário. O cientista de dados deve sempre considerar possíveis vieses de exposição e refletir sobre o significado prático dos achados antes de inferir relações causais ou sugerir recomendações.

**1. Padrões de Acidentes por Dia da Semana
Análise:**

Os acidentes de trânsito apresentam variações claras conforme o dia da semana, com aumento evidente em sextas, sábados e domingos. Esses picos sugerem influência de comportamento de deslocamento e lazer no risco de acidentes.

**Limitação/Exploração adicional:** Destacar se há mudanças também na gravidade dos acidentes aos fins de semana.

```python
plt.figure(figsize=(12, 5))
sns.countplot(
    x='dia_semana',
    data=df_consolidado,
    order=[
        'segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira',
        'sexta-feira', 'sábado', 'domingo'
    ]
)
plt.title('Distribuição de Acidentes por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Quantidade de Acidentes')
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
```

Abaixo apresento uma nova visualização que destaca, em percentual, a variação diária no número de acidentes segundo a gravidade. O gráfico mostra de forma clara quanto cada categoria aumentou ou diminuiu em relação a média semanal, permitindo identificar de modo detalhado os momentos de maior crescimento ou queda no risco de acidentes ao longo da semana.

```python
# Definimos um dicionário para associar cada categoria de gravidade de acidente a uma cor específica no gráfico,
# facilitando a visualização e interpretação dos resultados.
cores = {
    'COM VÍTIMAS FERIDAS': 'royalblue',
    'COM VÍTIMAS FATAIS': 'red',
    'SEM VÍTIMAS': 'green'
}

# Organizamos manualmente a ordem dos dias da semana para garantir que os resultados sigam a sequência lógica do calendário
dias = ['segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira', 'sexta-feira', 'sábado', 'domingo']

# Criamos a tabela grav_counts, que apresenta a contagem absoluta de acidentes para cada combinação de dia da semana e gravidade.
# Utilizamos groupby seguido de unstack para gerar uma tabela de contingência, reordenando pelos dias para manter a ordem correta.
grav_counts = (
    df_consolidado
    .groupby(['dia_semana','classificacao_acidente'])
    .size()
    .unstack()
    .reindex(dias)
    .fillna(0)
)

# Calculamos a média do número de acidentes em cada categoria ao longo da semana,
# ou seja, para cada tipo de gravidade, qual é o valor médio de acidentes considerando todos os dias.
media_geral_semana = grav_counts.mean()

# Para medir o quanto cada dia se distancia dessa média, calculamos a variação percentual do valor do dia
# em relação à média semanal (de toda a semana), para cada categoria.
# O resultado mostra, para cada dia, o percentual acima ou abaixo da média semanal.
grav_var_pct_media_geral = ((grav_counts - media_geral_semana) / media_geral_semana) * 100

# Plotamos o gráfico, onde cada linha representa a variação percentual diária em relação à média semanal
plt.figure(figsize=(14,7))
for categoria in grav_counts.columns:
    plt.plot(
        grav_counts.index,
        grav_var_pct_media_geral[categoria],
        marker='o',
        color=cores.get(categoria, 'gray'),
        label=categoria,
        linewidth=2
    )
    # Inserimos labels de variação percentual sobre cada ponto da curva
    for i, (x, y) in enumerate(zip(grav_var_pct_media_geral.index, grav_var_pct_media_geral[categoria])):
        label = f"{y:+.1f}%"
        plt.text(x, y, label, ha='center', va='bottom' if y>=0 else 'top',
                 fontsize=11, color=cores.get(categoria, 'gray'))
plt.axhline(0, color='black', linestyle='--', lw=1)
plt.title('Variação Percentual Diária em Relação à Média Semanal dos Acidentes por Gravidade', fontsize=15)
plt.xlabel('Dia da Semana')
plt.ylabel('Variação em relação à média semanal (%)')
plt.xticks(rotation=20, ha='right')
plt.legend(title='Classificação', fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

A análise dos dois gráficos revela padrões importantes quanto à distribuição dos acidentes ao longo da semana. No primeiro gráfico, que mostra o total de acidentes por dia, nota-se um crescimento gradual a partir do início da semana, atingindo seu pico no sábado (quase 50 mil acidentes), recuando levemente no domingo e apresentando os menores valores nas terças e quartas-feiras.

Já o segundo gráfico, que apresenta a variação percentual diária em relação à média semanal por gravidade, confirma que acidentes “sem vítimas” e “com vítimas feridas” seguem essa mesma tendência: maiores percentuais aos sábados (19,5% e 14,7%, respectivamente) e ligeira queda aos domingos. No entanto, os acidentes “com vítimas fatais” apresentam um padrão distinto, com um aumento mais intenso aos domingos (46,7% acima da média semanal), sinalizando um risco agravado nesse dia, apesar de o volume total de acidentes ser menor do que no sábado.

Esses resultados sugerem que, além do simples volume de acidentes, o domingo concentra acidentes potencialmente mais graves. Uma sugestão para aprofundar esse fenômeno seria a aplicação de técnicas avançadas, como machine learning ou análise de fatores contextuais, para identificar causas específicas e apoiar estratégias de prevenção focadas nos finais de semana, especialmente aos domingos.
